```python
## Types of inputs to plotting functions
# plotting functions expect numpy.array or numpy.ma.masked_array as input.
# Classes that are 'array-like' such as pandas data objects and numpy.matrix is best to convert to numpy.array objects prior to plotting.
# For example, to convert a pandas.DataFrame
a = pd.DataFrame(np.random.rand(4, 5), columns= list('abcde'))
print(a) #pandas DataFrame
print(type(a))
a_asarray = a.values
print(a_asarray) #numpy array
print(type(a_asarray))
```

```python
## The object-oriented interface and the pyplot interface
# There are essentially two ways to use Matplotlib:
# 1- Explicitly create figures and axes, and call methods on them (the "object-oriented (OO) style").

x = np.linspace(0, 2, 100) #create a numpy array of 100 values between 0 and 2
#print(x)
#print(type(x))
fig, ax = plt.subplots() #create a figure and axes
ax.plot(x, x, label='Linear') #plot data on axes between x and x (linear)
ax.plot(x, x**2, label='Quadratic') # plot data on axes between x and x squared (quadratic)
ax.plot(x, x**3, label='Cubic') ## plot data on axes between x and x cubic (cubic)
ax.set_xlabel('X label') # add an x-label
ax.set_ylabel('Y label') #add a y label
ax.set_title(" Simple plot") #add title
ax.legend() #add a legend
```

```python
# 2 - Rely on pyplot to automatically create and manage the figures and axes, and use plt.plot functions for plotting.
x = np.linspace(0, 2, 100) #create an array of 100 values between 0 and 2

plt.plot(x, x, label='Linear')
plt.plot(x, x**2, label='Quadratic')
plt.plot(x, x**3, label='Cubic')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title("Simple plot")
plt.legend()
```

```python
# making the same plots over and over again, but with different data sets, 
# leads to needing to write specialized functions to do the plotting.
def my_plotter(ax, data1, data2, param_dict):
    out = ax.plot(data1, data2, **param_dict)
    return out 

"""
function to make a graph

Parameters:
    ax: Axes
    data1: X data (array)
    data2: Y data (array)
    param_dict: Dictionary of kwargs for ax.plot
    
Returns: 
    out: list of artists
"""
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax1 = plt.subplots(1, 1)
my_plotter(ax1, data1, data2, {'marker': 'x'})
```

```python
fig, ax2 = plt.subplots(1, 1)
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data3, {'marker': 'x'})
my_plotter(ax2, data2, data4, {'marker': 'o'})
```
