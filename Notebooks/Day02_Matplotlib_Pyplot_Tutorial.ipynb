```python
##Intro to pyplot
# matplotlib.pyplot is a collection of functions that make matplotlib work like MATLAB. 
# Each pyplot function makes some change to a figure: 
# e.g., creates a figure, creates a plotting area, plots some lines, decorates the plot with labels, etc.
```

```python
#Generating visualizations with pyplot is very quick:

import matplotlib.pyplot as plt 
plt.plot([1, 2, 3, 4])
plt.ylabel('Some numbers')
plt.show()

# If you provide a single list or array to plot, matplotlib assumes it is a sequence of y values,
# and automatically generates the x values for you with the same length as y but starts with 0. Hence the x data are [0, 1, 2, 3].
```

```python
plt.plot([1, 2, 4, 6], [2, 4, 6, 9])
```

```python
## Formatting the style of your plot
# For every x, y pair of arguments, there is an optional third argument 
# which is the format string that indicates the color and line type of the plot.

plt.plot([1, 3, 4, 7], [2, 4, 6, 7], 'ro') # r means red and o is the point marker / format
plt.axis([0, 8, 0, 10]) # The axis function takes a list of [xmin, xmax, ymin, ymax], specifies the viewport of the axes
plt.show()
```

```python
# plotting several lines with different format styles in one function call using arrays.
import numpy as np
t = np.arange(0, 5, 0.2) # evenly sampled time at 200ms intervals

plt.plot(t, t, 'r--') ## red dashes, blue squares and green triangles
plt.plot( t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

```python
import numpy as np
t = np.arange(0, 5, 0.2) # evenly sampled time at 200ms intervals

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^') 
plt.show()
```

```python
## Plotting with keyword strings
#There are some instances where you have data in a format that lets you access particular variables with strings.
# For example, with numpy.recarray or pandas.DataFrame.

# Matplotlib allows you provide such an object with the data keyword argument. 
# If provided, then you may generate plots with the strings corresponding to these variables.

data = { 'a': np.arange(50),
       'c': np.random.randint(0, 50, 50),
       'd': np.random.randn(50)}
#print(data['a'])
#print(data['c'])
#print(data['d'])
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

#print(data['b'])
#print(data['d'])


plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
## Plotting with categorical variables¶
# It is also possible to create a plot using categorical variables. 
# Matplotlib allows you to pass categorical variables directly to many plotting functions. For example:

names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values, c='rgb')
plt.subplot(133)
plt.plot(names, values, )
plt.suptitle('Categorical plotting')
plt.show()
```

```python
## Controlling line properties

# Lines have many attributes that you can set: linewidth, dash style, antialiased, etc; 
# see matplotlib.lines.Line2D. There are several ways to set line properties

x = np.arange(10)
plt.plot(x, x**3, linewidth=10) #linewidth
```

```python
# Use the setter methods of a Line2D instance
y = np.arange(10)
line, = plt.plot(x, y**2, c='blue')
line.set_antialiased(False) #turn off antialiasing
```

```python
# Use setp: set multiple properties on a list of lines

x1 = np.arange(0, 10, 1)
y1 = np.arange(0, 20, 2)
x2 = np.arange(0, 30, 3)
y2 = np.arange(0, 40, 4)
lines = plt.plot(x1, y1, x2, y2)
plt.setp(lines, color='r', linewidth=2.0)
plt.show()

```
